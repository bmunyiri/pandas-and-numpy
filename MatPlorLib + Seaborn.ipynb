```python
# improve on the x axis
years = [2011,2012,2013,2014,2015]
yeld_apples = [0.567,0.685,0.745,0.846,0.967]
plt.plot(years,yeld_apples);
```

## Axis Lables

```python
plt.plot(years,yeld_apples)
plt.xlabel('Years')
plt.ylabel('Yeld (tons apples per hectare)' );
```

## Plotting Multipule Lines

```python
yeld_oranges = [0.234,0.456,0.524,0.559,0.599]
plt.plot(years,yeld_apples)
plt.plot(years,yeld_oranges)
plt.xlabel('Years')
plt.ylabel('Yeld (tons per hectare)' );
```

# Titile & Legends

```python
yeld_apples = [0.567,0.685,0.745,0.846,0.967]
yeld_oranges = [0.234,0.456,0.524,0.559,0.599]

plt.plot(years,yeld_apples)
plt.plot(years,yeld_oranges)

plt.xlabel('Years')
plt.ylabel('Yeld (tons per hectare)' );

plt.title("Apples And Oranges Farmming In Kenya")
plt.legend(['apples','oranges']);
```

# Line Styles & Markers

```python
yeld_apples = [0.567,0.685,0.745,0.846,0.967]
yeld_oranges = [0.234,0.456,0.524,0.559,0.599]

plt.plot(years,yeld_apples, marker='x',ls='--',c='r')
plt.plot(years,yeld_oranges, marker ='s')

plt.xlabel('Years')
plt.ylabel('Yeld (tons per hectare)' );

plt.title("Apples And Oranges Farmming In Kenya")
plt.legend(['apples','oranges']);
```
